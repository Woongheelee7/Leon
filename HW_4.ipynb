```python
image_data = Image.open(urlopen("https://raw.githubusercontent.com/changyaochen/MECE4520/master/data/leena.png"))
X=np.array(image_data)
plt.figure()
plt.imshow(image_data,cmap="gray")
plt.show()
```

```python
x_std = (X - np.mean(X,axis = 0))/ np.std(X,axis = 0)
X_std = np.round(x_std,5)
features = X_std.T 
covariance_matrix = np.cov(features)
eig_vals, eig_vecs = np.linalg.eig(covariance_matrix)
val_mat=np.diag(eig_vals)
U=eig_vecs.T[:,0:50]
X_sub=X_std
projected_X = X_sub.dot(U)
X_app=projected_X.dot(U.T)
a= np.sum((X_app[:,0]-projected_X[:,0])**2)/len(X_app[:,0])
b= np.sum((X_sub[:,0])**2)/len(X_sub[:,0])
print(a/b,U[0][0],X_std[129][129],X[128,128])
```
